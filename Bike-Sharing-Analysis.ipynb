# Training a (simple) Model

```python
# Let's take some simple scikit learn regression model
from sklearn import tree

# Let's prepare some of the features first. Since we are going to use regression and not a time serries method, 
# let's re-write the datetime to a feature called hour_of_day
from sklearn import preprocessing
data['hour_of_day'] = data['datetime'].apply(lambda i: i.hour)

# Let's start with taking only some columns and transforming from a DataFrame to a numpy array
train_data = data[['hour_of_day', 'workingday', 'temp']].values
train_labels = data[['count']].values

# Let scale the data set to have zero mean and unit variance, this makes sure variables with different orders of magnitude
# are treated equally
train_data = preprocessing.scale(train_data)


from sklearn import cross_validation
from sklearn.metrics import mean_squared_error
from prediction import rmsle
kf = cross_validation.KFold(len(train_data), n_folds=4, shuffle=True)

for train_index, test_index in kf:
    # Train the model
    clf = tree.DecisionTreeRegressor()
    clf.fit(train_data[train_index],train_labels[train_index])
    # Test it
    predicted = clf.predict(train_data[test_index])
    
    print "MSE:", math.sqrt(mean_squared_error(train_labels[test_index], predicted))
    print "RMSLE:", rmsle(train_labels[test_index], predicted)
```

```python
# A plot of one random day of the cross validation fold's results
import random

df = pd.DataFrame({'datetime': data['datetime'].values[test_index],'true': np.ravel(train_labels[test_index]),'predicted': np.ravel(predicted)})
index = random.randint(0,len(df))
df[index:index+48].plot(x='datetime')
```

```python
# Train again on all train data, so we have the best possible model for the real test set
clf.fit(train_data,train_labels)
```

# Using our Model 

```python
# Do the same transformations on the test data
test_data['hour_of_day'] = test_data['datetime'].apply(lambda i: i.hour)
transformed_test_data = test_data[['hour_of_day', 'workingday', 'temp']].values
transformed_test_data = preprocessing.scale(transformed_test_data)

# Predict all test data
predicted = clf.predict(transformed_test_data)
```

```python
# Write the output to a csv file (and manually upload to Kaggle)
output = pd.DataFrame()
output['datetime'] = test_data['datetime']
output['count'] = predicted

# Don't write row numbers, using index=False
output.to_csv('data/predicted.csv', index=False)
                   
# Does it look right?    
print output.head()
```
